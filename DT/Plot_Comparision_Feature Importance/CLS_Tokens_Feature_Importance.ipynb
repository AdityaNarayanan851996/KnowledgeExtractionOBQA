```python
%matplotlib inline
import pickle
import collections
import matplotlib.pyplot as plt
import numpy as np
```

```python
pickle_in = open("DT_75000_.pkl","rb")
model_DT = pickle.load(pickle_in)
```

```python
#print(model_DT.best_estimator_.feature_importances_)
```

```python
def top10FeaturePlotter(feature_importances,algo_name):
    fig = plt.figure()
    d = dict(enumerate(feature_importances))
    sorted_d = sorted(d.items(), key=lambda kv: kv[1],reverse=True)
    sorted_dict = collections.OrderedDict(sorted_d[0:10])
    x = list(sorted_dict.keys())
    y_pos = np.arange(len(x))
    y = list(sorted_dict.values())
    plt.bar(y_pos,y, align='center',label=str(algo_name))
    plt.xticks(y_pos, x)
    plt.title("Feature Importance vs Feature Index")
    plt.legend()
    plt.xlabel("Feature Index")
    plt.ylabel("Feature Importance Value")
    fig.savefig(str(algo_name)+"_top10_features.png")
    
```

```python
top10FeaturePlotter(model_DT.best_estimator_.feature_importances_,"Decision_Trees")
```
