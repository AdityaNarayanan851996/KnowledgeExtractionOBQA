```python
import matplotlib.pyplot as plt
```

```python
# With all_PCA dataset, best classifier had k=99
n_trainSamples = [10000,20000,50000,75000]

# Validation Accuracy
val_acc = [0.5207, 0.5309, 0.5498, 0.5498]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,val_acc,label='k=99' ,marker='o')

    plt.title("Validation Accuracy vs TrainData size")
    plt.legend()
    plt.xlabel("TrainData size")
    plt.ylabel("Accuracy")
    fig.savefig("plots/knn_pcaAll_Val_Accuracy_plot.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
