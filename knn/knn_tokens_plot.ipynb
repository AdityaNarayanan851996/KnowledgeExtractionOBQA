```python
import matplotlib.pyplot as plt
```

```python
# With tokens dataset, best classifier had k=949
# k=949, so train data size > 949
n_trainSamples = [10000,20000,50000,75000]

# Validation Accuracy
val_acc = [0.5619, 0.568, 0.5757, 0.5739]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,val_acc,label='k=949' ,marker='o')

    plt.title("Validation Accuracy vs TrainData size")
    plt.legend()
    plt.xlabel("TrainData size")
    plt.ylabel("Accuracy")
    fig.savefig("plots/knn_tokens_Val_Accuracy_plot.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
