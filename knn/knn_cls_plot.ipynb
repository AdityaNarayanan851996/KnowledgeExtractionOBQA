```python
import matplotlib.pyplot as plt
```

```python
# With ClS dataset, best classifier had k=99
n_trainSamples = [10000,20000,50000,75000]

# Validation Accuracy
val_acc = [0.5916, 0.6057, 0.6053, 0.6197]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,val_acc,label='k=99' ,marker='o')

    plt.title("Validation Accuracy vs TrainData size")
    plt.legend()
    plt.xlabel("TrainData size")
    plt.ylabel("Accuracy")
    fig.savefig("plots/knn_cls_Val_Accuracy_plot.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
