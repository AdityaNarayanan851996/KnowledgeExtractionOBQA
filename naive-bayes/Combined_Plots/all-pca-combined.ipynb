```python
import matplotlib.pyplot as plt
```

```python
# Validation Accuracy
GNB_val_acc = [0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]
MNB_val_acc = [0.5398, 0.5523, 0.5504, 0.5448, 0.547, 0.548, 0.5462]
BNB_val_acc = [0.4967, 0.4993, 0.4999, 0.5013, 0.5011, 0.5, 0.5]
n_trainSamples = [100,500,1000,10000,20000,50000,75000]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,GNB_val_acc,label='Gaussian')
    ax.plot(n_trainSamples,MNB_val_acc,label='Multinomial')
    ax.plot(n_trainSamples,BNB_val_acc,label='Bernoulli')
    plt.title("Validation Accuracy Scores vs TrainData")
    plt.legend()
    plt.xlabel("TrainData")
    plt.ylabel("Scores")
    fig.savefig("All_PCA_Validation_Accuracy_Comparision.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
