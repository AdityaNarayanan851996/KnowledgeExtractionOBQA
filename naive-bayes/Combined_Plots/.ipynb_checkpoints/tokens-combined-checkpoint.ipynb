```python
import matplotlib.pyplot as plt
```

```python
# Validation Accuracy
GNB_val_acc = [0.5058, 0.5045, 0.508, 0.5533, 0.5585, 0.5769, 0.5815]
MNB_val_acc = [0.5051, 0.5009, 0.5015, 0.5629, 0.5699, 0.5841, 0.5817]
BNB_val_acc = [0.5005, 0.5003, 0.5003, 0.5597, 0.5677, 0.5859, 0.5881]
n_trainSamples = [100,500,1000,10000,20000,50000,75000]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,GNB_val_acc,label='Gaussian')
    ax.plot(n_trainSamples,MNB_val_acc,label='Multinomial')
    ax.plot(n_trainSamples,BNB_val_acc,label='Bernoulli')
    plt.title("Validation Accuracy Scores vs TrainData")
    plt.legend()
    plt.xlabel("TrainData")
    plt.ylabel("Scores")
    fig.savefig("Tokens_Validation_Accuracy_Comparision.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
