```python
import matplotlib.pyplot as plt
```

```python
# Validation Accuracy
GNB_val_acc = [0.5, 0.5299, 0.5343, 0.5137, 0.5166, 0.5, 0.5]
MNB_val_acc = [0.5239, 0.5155, 0.5293, 0.5445, 0.5343, 0.5465, 0.5479]
BNB_val_acc = [0.5011, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5]
n_trainSamples = [100,500,1000,10000,20000,50000,75000]
```

```python
def valAccuracyVisualization():
    ''' Plots the data points with the centroids
    '''
    fig = plt.figure()
    ax = plt.axes()
    ax.plot(n_trainSamples,GNB_val_acc,label='Gaussian')
    ax.plot(n_trainSamples,MNB_val_acc,label='Multinomial')
    ax.plot(n_trainSamples,BNB_val_acc,label='Bernoulli')
    plt.title("Validation Accuracy Scores vs TrainData")
    plt.legend()
    plt.xlabel("TrainData")
    plt.ylabel("Scores")
    fig.savefig("All_Validation_Accuracy_Comparision.png")
    #plt.show()
```

```python
valAccuracyVisualization()
```
